```python
def decay_step(text):
    letters_in_text = set(text.lower()) & set(LETTER_FREQUENCIES.keys())
    if len(letters_in_text) == 0:
        return None
    letter_frequencies = { k:LETTER_FREQUENCIES[k] for k in letters_in_text } 
    letter_frequencies_sum = sum(letter_frequencies.values())
    letter_probabilities = { k:v/letter_frequencies_sum for (k, v) in letter_frequencies.items() }
    threshold = 0.0;
    letter_thresholds = []
    for letter in letters_in_text:
        letter_thresholds.append((letter, threshold))
        threshold += letter_probabilities[letter]
    rand = random.random()
    chosen_letter = None;
    for (letter, threshold) in letter_thresholds:
        if threshold < rand:
            chosen_letter = letter
        else:
            break
    replacements = {m.pos+m.start():m.group(0) for m in re.finditer(chosen_letter, text, re.IGNORECASE)}
    replace_index = random.choice(list(replacements.keys()))
    replacements[replace_index] = ""
    output = copy(text);
    offset = 0
    for (ix, replacement) in replacements.items():
       output = output[:ix+offset] + replacement + output[ix+offset + 1:]
       if replacement == "":
           offset = -1
    return output
```
